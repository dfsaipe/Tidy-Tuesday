```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

munros = pd.read_csv('Data/scottish_munros.csv', encoding="cp1252")
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(munros.isna(), cbar=False, cmap="viridis")
plt.title("NaN Heatmap")
plt.show()
```

```python
munros.dropna(subset=['Name'], inplace=True)
```

```python
dist_matrix = np.zeros((len(munros), len(munros)))
remote = pd.DataFrame(index=munros['Name'], columns=['Remote'])

for i in range(len(munros)):
    for j in range(len(munros)):
        dist_matrix[i,j] = np.sqrt((munros['xcoord'][i] - munros['xcoord'][j])**2 
                                   + (munros['ycoord'][i] - munros['ycoord'][j])**2)
        

        
for i in range(len(munros)):
    remote.iloc[i,0] = dist_matrix[i,:].mean()
        
remote.sort_values(by='Remote', ascending=False, inplace=True)
remote.head()
```
